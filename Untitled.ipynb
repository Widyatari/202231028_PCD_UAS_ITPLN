```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'ProjectPCD.jpg'
```

# Muat gambar

```python
image = cv2.imread(image_path)
```

# Periksa apakah gambar berhasil dimuat

```python
if image is None:
    print("Gambar tidak ditemukan di jalur:", image_path)
else:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Konversi gambar ke RGB
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Konversi gambar ke skala abu-abu
    edges = cv2.Canny(gray_image, 100, 200) # Terapkan deteksi tepi Canny
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  # Temukan kontur
    image_contours = image_rgb.copy() # Salin gambar asli untuk menggambar kontur
    cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 3)
```

# Buat Plot Untuk menampilkan Gambar

```python
# OUTPUT 1
plt.figure(figsize=(15, 10))
```

```python
plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
```

```python
# OUTPUT 2
plt.subplot(2, 3, 4)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(image_contours)
plt.title('Contours Detection')
plt.axis('off')

plt.show()
```
